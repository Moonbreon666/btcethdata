```python
import pandas as pd

# Define file paths
btc = r"/Users/user/Downloads/archive (2)/BTC-USD From 2014 To Dec-2024.csv"
eth = r"/Users/user/Downloads/archive (2)/ETH-USD From 2017 To Dec-2024.csv"

# Load the CSV files into DataFrames
btc_data = pd.read_csv(btc)
eth_data = pd.read_csv(eth)

# Ensure the 'Date' column is in datetime format
btc_data['Date'] = pd.to_datetime(btc_data['Date'])
eth_data['Date'] = pd.to_datetime(eth_data['Date'])

# Filter both datasets to include only data from 2017 onwards
btc_data_cleaned = btc_data[btc_data['Date'] >= '2017-01-01']
eth_data_cleaned = eth_data[eth_data['Date'] >= '2017-01-01']  # ETH data already starts from 2017

# Optional: Align columns if there are differences
# Ensure common_columns is converted to a list
common_columns = list(set(btc_data_cleaned.columns).intersection(set(eth_data_cleaned.columns)))

# Use the list of columns to index the DataFrames
btc_data_cleaned = btc_data_cleaned[common_columns]
eth_data_cleaned = eth_data_cleaned[common_columns]\


# Inspect the cleaned datasets
print("Cleaned BTC Data - First 5 Rows:")
print(btc_data_cleaned.head())

print("\nCleaned ETH Data - First 5 Rows:")
print(eth_data_cleaned.head())

# Save cleaned datasets back to CSV (optional)
btc_data_cleaned.to_csv("btc_data_cleaned.csv", index=False)
eth_data_cleaned.to_csv("eth_data_cleaned.csv", index=False)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Define file paths
btc = r"/Users/user/Downloads/archive (2)/BTC-USD From 2014 To Dec-2024.csv"
eth = r"/Users/user/Downloads/archive (2)/ETH-USD From 2017 To Dec-2024.csv"

# Load the CSV files into DataFrames
btc_data = pd.read_csv(btc)
eth_data = pd.read_csv(eth)

# Ensure the 'Date' column is in datetime format
btc_data['Date'] = pd.to_datetime(btc_data['Date'])
eth_data['Date'] = pd.to_datetime(eth_data['Date'])

# Filter both datasets to include only data from 2017 onwards
btc_data_cleaned = btc_data[btc_data['Date'] >= '2017-01-01']
eth_data_cleaned = eth_data  # ETH data already starts from 2017

# Plot BTC data with Volume
fig, ax1 = plt.subplots(figsize=(12, 6))

# Bitcoin Closing Price
ax1.plot(btc_data_cleaned['Date'], btc_data_cleaned['Close'], label='Bitcoin Close Price', color='blue')
ax1.set_xlabel("Date")
ax1.set_ylabel("Closing Price (USD)", color='blue')
ax1.tick_params(axis='y', labelcolor='blue')
ax1.grid(True)

# Bitcoin Volume on a secondary y-axis
ax2 = ax1.twinx()
ax2.plot(btc_data_cleaned['Date'], btc_data_cleaned['Volume'], label='Bitcoin Volume', color='gray', alpha=0.6)
ax2.set_ylabel("Volume", color='gray')
ax2.tick_params(axis='y', labelcolor='gray')

fig.suptitle("Bitcoin Price and Volume Over Time")
fig.tight_layout()
plt.show()

# Plot ETH data with Volume
fig, ax1 = plt.subplots(figsize=(12, 6))

# Ethereum Closing Price
ax1.plot(eth_data_cleaned['Date'], eth_data_cleaned['Close'], label='Ethereum Close Price', color='orange')
ax1.set_xlabel("Date")
ax1.set_ylabel("Closing Price (USD)", color='orange')
ax1.tick_params(axis='y', labelcolor='orange')
ax1.grid(True)

# Ethereum Volume on a secondary y-axis
ax2 = ax1.twinx()
ax2.plot(eth_data_cleaned['Date'], eth_data_cleaned['Volume'], label='Ethereum Volume', color='gray', alpha=0.6)
ax2.set_ylabel("Volume", color='gray')
ax2.tick_params(axis='y', labelcolor='gray')

fig.suptitle("Ethereum Price and Volume Over Time")
fig.tight_layout()
plt.show()
```
